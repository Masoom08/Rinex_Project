```python
#Do data ingestion
import pandas as pd
df = pd.read_csv('/content/weatherAUS.csv')
df.head()
```

#preprocessing

```python
#Handling missing values
df.isnull().sum()
```

```python
df = df.fillna(df.isnull().sum())
```

```python
pd.get_dummies(df, columns=['RainToday','RainTomorrow'])
```

```python
mapping_dict = {"Yes": 1,"No":0,"NaN":0,"Na":0}
df['RainToday'] = df['RainToday'].map(mapping_dict)
df['RainTomorrow'] = df['RainTomorrow'].map(mapping_dict)
```

```python
#Feature scaling/normalization
df['RainToday'] = (df['RainToday'] - df['RainToday'].mean()) / df['RainToday'].std()
```

```python
df['RainToday'] = (df['RainToday'] - df['RainToday'].min()) / (df['RainToday'].max() - df['RainToday'].min())
```

```python
df['RainTomorrow'] = (df['RainTomorrow'] - df['RainTomorrow'].mean()) / df['RainTomorrow'].std()
```

```python
df['RainTomorrow'] = (df['RainTomorrow'] - df['RainTomorrow'].min()) / (df['RainTomorrow'].max() - df['RainTomorrow'].min())
```
