```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load dataset
df = pd.read_csv('/content/Lemonade.csv')
df.head()
```

```python
# Select input features and output target
X = df[['Temperature', 'Rainfall', 'Flyers']]
y = df['Sales']
X
y
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict for testing data
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

```python
# Plot the data and the best-fit line
fig, ax = plt.subplots()
ax.scatter(X_test['Temperature'], y_test, color='black')
ax.plot(X_test['Temperature'], y_pred, color='blue', linewidth=3)
ax.set_xlabel('Temperature')
ax.set_ylabel('Sales')
plt.show()
```
